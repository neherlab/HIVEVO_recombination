```python
# Adds link to the scripts folder
import sys
import os
sys.path.append("../../scripts/")
import matplotlib.pyplot as plt
import numpy as np

from divergence import load_divergence_dict, WH_evo_rate
```

# Divergence to founder

```python
time_average = np.arange(0, 3100, 100)
divergence_dict = load_divergence_dict("../../divergence_dict_founder")

colors = ["C0", "C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9"]
fontsize = 16
```

```python
plt.figure(figsize=(14, 10))
for ii, region in enumerate(["env","pol","gag"]):
    plt.plot(time_average, divergence_dict[region]["consensus"]["all"], '-', color=colors[ii], label=region)
    plt.plot(time_average, divergence_dict[region]["non_consensus"]["all"], '--', color=colors[ii])
    plt.plot(time_average, divergence_dict[region]["all"]["all"], ':', color=colors[ii])
plt.plot([0], [0], 'k-', label="Consensus")
plt.plot([0], [0], 'k--', label="Non-consensus")
plt.plot([0], [0], 'k:', label="All")
plt.grid()
plt.xlabel("Time since infection [days]", fontsize=fontsize)
plt.ylabel("Divergence", fontsize=fontsize)
plt.legend(fontsize=fontsize)
plt.show()
```

# Divergence founder VS consensus

```python
time_average = np.arange(0, 3100, 100)
divergence_dict_founder = load_divergence_dict("../../divergence_dict_founder")
divergence_dict_consensus = load_divergence_dict("../../divergence_dict_consensus")

plt.figure(figsize=(14, 10))
for ii, region in enumerate(["env","pol","gag"]):
    plt.plot(time_average, divergence_dict_founder[region]["all"]["all"], '-', color=colors[ii])
    plt.plot(time_average, divergence_dict_consensus[region]["all"]["all"], '--', color=colors[ii])
plt.plot([0], [0], 'k-', label="Founder")
plt.plot([0], [0], 'k--', label="Consensus")
plt.grid()
plt.xlabel("Time since infection [days]", fontsize=fontsize)
plt.ylabel("Divergence", fontsize=fontsize)
plt.legend(fontsize=fontsize)
plt.show()

plt.figure(figsize=(14, 10))
for ii, region in enumerate(["env","pol","gag"]):
    plt.plot(time_average, divergence_dict_founder[region]["all"]["all"], '-', color=colors[ii])
    plt.plot(time_average, divergence_dict_consensus[region]["all"]["all"]-divergence_dict_consensus[region]["all"]["all"][0], '--', color=colors[ii])
plt.plot([0], [0], 'k-', label="Founder")
plt.plot([0], [0], 'k--', label="Consensus")
plt.grid()
plt.xlabel("Time since infection [days]", fontsize=fontsize)
plt.ylabel("Divergence", fontsize=fontsize)
plt.legend(fontsize=fontsize)
plt.show()
```
